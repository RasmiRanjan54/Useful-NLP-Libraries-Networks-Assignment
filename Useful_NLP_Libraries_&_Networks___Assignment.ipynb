# Question 5: What is the key difference between LSTM and GRU networks in NLP applications?

- LSTM and GRU are advanced RNN variants that solve vanishing gradients, but differ fundamentally in gating mechanisms and efficiency. Key difference: LSTM uses 3 gates + separate cell state (4x parameters); GRU uses 2 gates + merged state (~75% LSTM parameters).

* Architectural Differences
1. LSTM
2. GRU

 Structure в–¶
1. More complex
2. Simpler

Number of Gates в–¶
1. 3 gates
2. 2 gates

Memory Cell в–¶
1. Has a separate memory cell
2. No separate memory cell

Training Speed в–¶
1. Slower
2. Faster

Parameters в–¶
1. More parameters
2. Fewer parameters

Performance в–¶
1. Good for long sequences
2. Good for moderate sequences

* LSTM Architecture

LSTM uses three gates to control information flow:

Forget Gate вҖ“ decides what information to discard

Input Gate вҖ“ decides what new information to store

Output Gate вҖ“ decides what information to send to the output

Because of these gates, LSTM can remember long-term dependencies in text sequences.

Example use:

Language translation

Text generation

Speech recognition

* GRU Architecture

GRU simplifies the LSTM structure by using two gates:

Update Gate вҖ“ controls how much past information to keep

Reset Gate вҖ“ decides how to combine new input with past memory

GRU merges the cell state and hidden state, which reduces complexity and speeds up training.

```python
# Question 7: Given the sample paragraph below, perform string tokenization and frequency distribution using Python and NLTK:

"""Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical."""

!pip install nltk

import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
from nltk.probability import FreqDist

nltk.download('punkt')
nltk.download('punkt_tab')

text = """Natural Language Processing (NLP) is a fascinating field that combines linguistics,
computer science, and artificial intelligence. It enables machines to understand,
interpret, and generate human language. Applications of NLP include chatbots,
sentiment analysis, and machine translation. As technology advances, the role of NLP
in modern solutions is becoming increasingly critical."""

tokens = word_tokenize(text)
print("Tokens:")
print(tokens)

freq_dist = FreqDist(tokens)

print("\nFrequency Distribution:")

for word, freq in freq_dist.items():
    print(f"{word}: {freq}")
```

```python
# Question 8: Implement a basic LSTM model in Keras for a text classification task using the following dummy dataset. Your model should classify sentences as either positive (1) or negative (0).
# Dataset
"""texts = [
вҖңI love this projectвҖқ, #Positive
вҖңThis is an amazing experienceвҖқ, #Positive
вҖңI hate waiting in lineвҖқ, #Negative
вҖңThis is the worst serviceвҖқ, #Negative
вҖңAbsolutely fantastic!вҖқ #Positive
]
labels = [1, 1, 0, 0, 1]"""

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt

# 1. Dataset
texts = [
    "I love this project",
    "This is an amazing experience",
    "I hate waiting in line",
    "This is the worst service",
    "Absolutely fantastic!"
]
labels = np.array([1, 1, 0, 0, 1])

print("Dataset:")
for text, label in zip(texts, labels):
    sentiment = "POSITIVE" if label == 1 else "NEGATIVE"
    print(f"  '{text}' вҶ’ {sentiment}")

# 2. TEXT PREPROCESSING
# Tokenization & Vocabulary
tokenizer = Tokenizer(num_words=100, oov_token="<OOV>")
tokenizer.fit_on_texts(texts)
vocab_size = len(tokenizer.word_index) + 1
print(f"\nVocab size: {vocab_size}")
print("Word index:", tokenizer.word_index)

# Convert to sequences
sequences = tokenizer.texts_to_sequences(texts)
print("\nSequences:", sequences)

# Padding (fixed length)
max_length = 10
padded_sequences = pad_sequences(sequences, maxlen=max_length, padding='post', truncating='post')
print("\nPadded sequences:\n", padded_sequences)

# 3. MODEL ARCHITECTURE
model = Sequential([
    Embedding(vocab_size, 32, input_length=max_length),  # Embeddings
    LSTM(64, dropout=0.3, recurrent_dropout=0.3),        # LSTM layer
    Dropout(0.5),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')                       # Binary classification
])

model.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()

# 4. TRAINING
print("\nTRAINING...")
history = model.fit(
    padded_sequences, labels,
    epochs=20,
    batch_size=8,
    validation_split=0.2,  # 1 sample validation
    verbose=1
)

# 5. EVALUATION & PREDICTIONS
print("\nPREDICTIONS:")
for i, seq in enumerate(padded_sequences):
    pred = model.predict(seq.reshape(1, -1), verbose=0)[0][0]
    pred_label = 1 if pred > 0.5 else 0
    actual = labels[i]
    print(f"Text: '{texts[i]}'")
    print(f"  Actual: {actual}, Predicted: {pred_label:.0f} (prob: {pred:.3f})")

print("\nFinal Accuracy:", max(history.history['val_accuracy']))

```

```python
# Question 10: You are working on a chatbot for a mental health platform. Explain how you would leverage LSTM or GRU networks along with libraries like spaCy or Stanford NLP to understand and respond to user input effectively. Detail your architecture, data preprocessing pipeline, and any ethical considerations.

# Install libraries
!pip install spacy tensorflow
!python -m spacy download en_core_web_sm

import spacy
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Example dataset
texts = [
    "I feel very sad today",
    "I am extremely happy",
    "I feel stressed about my exams",
    "Life is wonderful",
    "I feel lonely"
]

labels = np.array([0, 1, 0, 1, 0])  # 0 = negative emotion, 1 = positive emotion

# Preprocessing with spaCy
clean_texts = []
for text in texts:
    doc = nlp(text.lower())
    tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]
    clean_texts.append(" ".join(tokens))

print("Cleaned Texts:", clean_texts)

# Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(clean_texts)

seq = tokenizer.texts_to_sequences(clean_texts)

# Padding
X = pad_sequences(seq, maxlen=5)
y = labels

# Build LSTM model
model = Sequential()
model.add(Embedding(input_dim=50, output_dim=8, input_length=5))
model.add(LSTM(16))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train model
model.fit(X, y, epochs=20, verbose=0)

# Test input
test = ["I feel very anxious"]
doc = nlp(test[0].lower())
tokens = [t.lemma_ for t in doc if not t.is_stop and t.is_alpha]
clean_test = " ".join(tokens)

seq_test = tokenizer.texts_to_sequences([clean_test])
pad_test = pad_sequences(seq_test, maxlen=5)

prediction = model.predict(pad_test)

print("\nUser Input:", test[0])
print("Prediction (probability of positive):", prediction)
```
